## Handel Missing Values ##

**This step ensures that the dataset is clean and consistent for analysis.**  
- **Converts the 'last_review' column to proper datetime format using `errors='coerce'` to handle invalid entries.**  
- **Fills missing values in 'reviews_per_month' with 0, assuming no reviews.**  
- **Drops rows with missing values in critical columns such as 'name', 'host_name', 'neighbourhood_group', etc.**  
- **Resets the index after dropping rows to maintain a clean DataFrame.**  
**These actions prepare the dataset for accurate analysis and visualizations.**


```python
# Convert 'last_review' column to datetime format
# This helps in performing date-based analysis and avoids errors due to incorrect formats.
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

# Fill missing values
df.fillna({
    'reviews_per_month': 0,
    'last review': df['last review'].min()
}, inplace=True)

# Drop records with missing 'name' or 'host name'
df.dropna(subset=['NAME', 'host name'], inplace=True)
```

```python
df.isnull().sum()
```

## Correct Data Types ##

**Ensure that all columns have the correct data types**

```python
# Remove dollar signs and convert to float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$]', '', regex=True).astype(float)
```

## Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

## Confirm Data Cleaning

```python
df.info()
```

**Drop columns with too many missing values or not useful for analysis** 

**- license`: Only 2 non-null values, so it's not helpful**  

**- house_rules`: Mostly empty or irrelevant for core analysis** 

```python
df.drop(columns=['license', 'house_rules'], errors='ignore')
```

## Visualization ##




## Distribution of Prices ##

**Plot the distribution of listing prices**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color='red')  # Set histogram color
plt.title('Distribution of Listing Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

**ðŸ“Š Interpretation: Distribution of Listing Prices**  
The histogram shows a fairly even distribution of listing prices across different price ranges, indicating no particular concentration of listings in any specific range.  

The KDE (Kernel Density Estimate) line helps visualize this even spread more clearly, confirming that the dataset contains listings with a wide variety of prices.


## Room Type Analysis ##

**Analyze the Distribution of Room Types**  

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='room type', data=df, color='hotpink')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

**ðŸ›ï¸ Room Type Distribution**  

**This bar chart shows the number of listings available for each room type. The majority of listings are for 'Entire home/apt', followed by 'Private room'. Shared and hotel rooms are less common.**  

**This suggests that most Airbnb users either rent out or prefer full private units over shared accommodations.**


## Neighborhood Analysis ## 

**Examines how listings are distributed across different neighborhoods.**


```python
plt.figure(figsize=(12, 8))
sns.countplot(
    y='neighbourhood group',
    data=df,
    color='lightgreen',
    order=df['neighbourhood group'].value_counts().index
)
plt.title('Number of Listings by Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

**The countplot shows a clear distribution of the number of listings across different neighborhood groups.  
Manhattan and Brooklyn dominate the listings, suggesting they are prime locations for Airbnb.  
Queens, Bronx, and Staten Island have fewer listings, indicating less availability or popularity in those areas.**


## Price vs. Room Type ##

**visualizes the relationship between listing price and room type.**


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

**Price vs. Room Type**  

**The boxplot provides a detailed view of how prices vary across different room types in the Airbnb dataset.  
It shows that while 'Shared room' tends to have lower prices, 'Private room', 'Entire home/apt', and 'Hotel room' have higher and more varied price ranges.  
This visualization helps in understanding the pricing dynamics for different types of accommodations on Airbnb.**


## Reviews Over Time ##

**Plot the number of reviews changes over time.**

```python
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
```

**The line plot provides a clear visualization of the number of reviews over time.  
It helps identify trends and patterns in review activity, such as periods of high or low engagement.  
This information is useful for understanding user behavior and the popularity of Airbnb listings throughout the year.  
Significant spikes or drops in reviews may warrant further investigation to uncover potential causes such as policy changes, market conditions, or external events.**
